```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('D:\Machine Learning\Coding Elements\Machine Learning\data/titanic.csv')
```

```python
df.head(20)
```

```python
df.keys()
```

```python
df.drop(['alive', 'deck','embark_town', 'who','embarked','fare'], axis = 1, inplace = True )
```

```python
df.head()
```

```python
dummy = pd.get_dummies(df['sex'])
```

```python
dummy.head()
```

```python
df = pd.concat([df, dummy], axis = 1)
```

```python
df.head()
```

```python
df = df.drop('sex', axis = 1)
df.head()
```

```python
dum = pd.get_dummies(df['adult_male'])
dum.head()
```

```python
df = pd.concat([df, dum], axis = 1)
df.head()
```

```python
df = df.drop('adult_male', axis = 1)
df.head()
```

```python
dumm = pd.get_dummies(df['class'])
df = pd.concat([df, dumm], axis = 1)
df = df.drop('class', axis = 1)
df.head()
```

```python
dummy_v = pd.get_dummies(df['alone'])
df = pd.concat([df, dummy_v], axis = 1)
df = df.drop('alone', axis = 1)
df.head()
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
df.head()
```

```python
df = df.replace(to_replace = ['NaN'], value = '')
```

```python
df['age'].value_counts().head()
```

```python
df.isnull().values.any()
```

```python
df = df.fillna(method = 'ffill')
df.tail()
```

```python
df.isnull().values.any()
```

```python
X = df.iloc[:,1:]
X.head()
```

```python
y = df.iloc[:,0]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```
